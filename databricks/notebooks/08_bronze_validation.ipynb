```python
tables = spark.sql("SHOW TABLES IN fraud_bronze")

display(tables)
```

```python
tables = [

    "bronze_transactions",
    "bronze_identity",
    "bronze_customer_master",
    "bronze_merchant_master",
    "bronze_country_risk",
    "bronze_exchange_rates",
    "bronze_ofac",
    "bronze_pipeline_config"

]

for table in tables:
    cnt = spark.sql(f"SELECT COUNT(*) AS cnt FROM fraud_bronze.{table}").collect()[0]["cnt"]
    print(f"{table:35} {cnt}")
```

```python
spark.sql("""

SELECT
customer_id,
COUNT(*) AS cnt

FROM fraud_bronze.bronze_customer_master

GROUP BY customer_id

HAVING COUNT(*) > 1

""").display()
```

```python
spark.sql("""

SELECT
merchant_id,
COUNT(*) AS cnt

FROM fraud_bronze.bronze_merchant_master

GROUP BY merchant_id

HAVING COUNT(*) > 1

""").display()
```
